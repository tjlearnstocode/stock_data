```python
shortInterestDf = pd.read_csv(
    "stock-loan-bal-sec-data.csv",
    header=0,
    encoding='utf-8'
)

directory_dataframe = pd.read_fwf("dlpDownload.txt",
header=None
)

directory_dataframe.head()
```

```python
shortInterestDf['Symbol'].str.strip()
tickers = []
tickers = shortInterestDf['Symbol'].str.strip()
#shortInterestDf = shortInterestDf.set_index('Symbol')
shortInterestDf.head()
ticker_list = list(tickers)
```

```python
yDf = pd.DataFrame(columns = [
'ticker',
'previousClose',
'volume',
'marketCap',
'floatShares',
'sharesOutstanding',
'sharesShort',
'sharesShortPriorMonth',
'sharesPercentSharesOut',
'shortPercentOfFloat'
])

aTickers = yf.Tickers(ticker_list)

for t in tickers:
    
    yDict = aTickers.tickers[t].info

    symbol=yDict['symbol']
    try:

        symbol=yDict['symbol']
        previousClose=yDict['previousClose']
        volume=yDict['volume']
        
        try:
            marketCap=yDict['marketCap']
            floatShares=yDict['floatShares']
            sharesOutstanding=yDict['sharesOutstanding']
            sharesShort=yDict['sharesShort']
            sharesShortPriorMonth=yDict['sharesShortPriorMonth']
            sharesPercentSharesOut=yDict['sharesPercentSharesOut']
            shortPercentOfFloat=yDict['shortPercentOfFloat']
            
            
            testlist = [
                symbol,
                previousClose,
                volume,
                marketCap,
                floatShares,
                sharesOutstanding,
                sharesShort,
                sharesShortPriorMonth,
                sharesPercentSharesOut,
                shortPercentOfFloat
            ]
        except:
            testlist = [
                symbol,
                previousClose,
                volume,
                "NA",
                "NA",
                "NA",
                "NA",
                "NA",
                "NA",
                "NA"
            ]
    except:
        testlist = [
            symbol,
            "NA",
            "NA",
            "NA",
            "NA",
            "NA",
            "NA",
            "NA",
            "NA",
            "NA"
        ]

    yDf.loc[len(yDf)] = testlist
```
